## Fitting the original LPPLS model

Like the LPPL model, the LPPLS model assumes that asset prices are driven by a complex system of market participants with heterogeneous beliefs, expectations, and trading strategies. In this framework, bubbles are characterized by a self-reinforcing feedback loop between bullish expectations and actual price increases, leading to an exponential rise in prices that eventually become unsustainable.

The LPPLS model incorporates several key assumptions:

1. The market is efficient in the long run, but can be subject to temporary bubbles or crashes due to irrational behavior or exogenous shocks.

2. The price of an asset is driven by a combination of fundamental factors, such as market capitalization, network activity, or social sentiment, and speculative factors, such as herd behavior or momentum trading.

3. Bubbles are characterized by a log-periodic power law pattern, where the price increases with a non-linear power function of time, and the frequency of oscillation increases as the critical time of the bubble approaches.

4. The LPPLS model can be used to predict the timing and magnitude of a bubble's collapse, based on the estimated critical time, the LPPL exponent, and the amplitude of the oscillations, as well as the impact of generalized error and jump-diffusion processes.

The LPPLS model uses a power-law function to describe the growth of the market, and a logarithmic function to describe the oscillations in the market. The model is characterized by several parameters, including the critical time, the critical amplitude, and the critical exponent, as well as the parameters that capture the impact of generalized error and jump-diffusion processes.

The LPPLS model can be represented by the following equation:

$$\ln(P(t))=A+B(t_c-t)^m+C(t_c-t)^m\cos(\omega\ln(t_c-t)+\phi)+D\sum_{i=1}^{n}e^{-\beta(t-t_i)}+ J(t) +\sigma_t \cdot z_t + \epsilon_t$$

The LPPLS (Log-Periodic Power Law Singularity) model is a mathematical equation used to analyze the behavior of financial markets, particularly during speculative bubbles. It is a non-linear model that aims to capture the dynamics of price movements in financial markets by modeling the log of the asset price, denoted as $\ln(P(t))$, as a function of time, $t$. The LPPLS model can be broken down into several components, which are as follows:

1. **Logarithmic trend component**: $A$. This component represents the long-term trend in the price of the asset and is modeled using a constant term. It captures the overall upward or downward movement of the price, irrespective of the short-term fluctuations.

2. **Power-law component**: $B(t_c-t)^m$. This component represents the power-law growth rate of the asset price leading up to the bubble peak. $B$ is a scaling factor that determines the strength of the power-law growth rate, while $t_c$ represents the time of the expected crash or bubble peak, and $m$ is a parameter that determines the curvature of the power-law growth rate.

3. **Log-periodic component**: $C(t_c-t)^m\cos(\omega\ln(t_c-t)+\phi)$. This component represents the periodic oscillations that occur within the bubble. The term $C$ is a scaling factor that determines the amplitude of the oscillations, while $\omega$ and $\phi$ are parameters that control the frequency and phase of the oscillations, respectively.

4. **Correction factor component**: $D\sum_{i=1}^{n}e^{-\beta(t-t_i)}$. This component represents the correction factor that adjusts the power-law growth rate and log-periodic oscillations to account for the effects of news and other exogenous factors. $D$ is a scaling factor that determines the strength of the correction factor, while $\beta$ represents the decay rate of the correction factor, and $t_i$ denotes the time of each exogenous event.

5. **Jump term component**: $J_t$ This component captures sudden, large price movements.

6. **Volatility term component**: $\sigma_t \cdot z_t$. This component captures the estimated GARCH volatility of the asset at time $t$ and is modeled as the product of a random variable $z_t$ drawn from a standard normal distribution and the estimated volatility $\sigma_t$.

7. **Error term component**: $\epsilon_t$. This component represents the random fluctuations around the model predictions.

For the sake of simplification and for the next sections, let's call a new variable, $y(t)$ our model:

$$y(t) = \ln(P(t))$$

### The critical time

In the LPPLS (Log-Periodic Power Law Singularity) model, $t_c$ represents the expected time of the crash or the bubble peak. The definition of $t_c$ in the LPPLS model is as follows:

$$t_c = t_m - \ln(\frac{\frac{\omega}{2 \pi}}{\omega})$$

where $t_m$ is the estimated time of the maximum of the log-periodic power law, and $\omega$ is the frequency of oscillation of the log-periodic power law.

This equation provides an estimate of the time of the bubble peak or crash, which is a key parameter in the LPPLS model. The value of $t_c$ can be used to predict the timing of a potential market crash or bubble peak, and to estimate the probability of such an event occurring based on the parameters of the LPPLS model. 

## Estimation

### The LPPL model estimation

In the LPPLS model, the critical exponent, denoted by $m$, is a parameter that determines the curvature of the power-law growth rate of the asset price leading up to the bubble peak. It is one of the key parameters of the LPPLS model and can be estimated using nonlinear regression techniques.

The LPPLS exponent can be interpreted as the degree of nonlinearity of the power-law growth rate. When $m > 1$, the power-law growth rate accelerates as the bubble peak approaches, while for $m < 1$, the power-law growth rate decelerates. When $m = 1$, the power-law growth rate is linear.

To estimate the LPPLS exponent, we can use the following equation:

$$\ln(P(t))=A+B(t_c-t)^m+C(t_c-t)^m\cos(\omega\ln(t_c-t)+\phi)+D\sum_{i=1}^{n}e^{-\beta(t-t_i)}$$

### The critical time $t_c$

The expression for t_c used in the LPPLS model is based on the assumption that the log-periodic power law pattern of the asset price is symmetric around the critical time. The critical time is assumed to lie in the interval $[T_{min}, T_{max}]$, where $T_{min}$ and $T_{max}$ are the minimum and maximum values of time in the training data set X_train.

The expression for t_c can be derived by setting the derivative of the log-periodic power law function with respect to time equal to zero at the critical time. This yields the following equation:

$$\ln(P(t_c))=A+B(t_c-t_c)^m+C(t_c-t_c)^m\cos(\omega\ln(t_c-t_c)+\phi)+D\sum_{i=1}^{n}e^{-\beta(t_c-t_i)}$$

Simplifying this equation gives:

$$\ln(P(t_c))=A+D\sum_{i=1}^{n}e^{-\beta(t_c-t_i)}$$

Taking the derivative of both sides of this equation with respect to t_c gives:


$$\frac{\ln(P(t_c))}{dt_c}= -\beta D\sum_{i=1}^{n}e^{-\beta(t_c-t_i)}$$

Setting this expression equal to zero and solving for t_c gives:

$$t_c = T_{max} - L^{\frac{1}{m-1}}(T_{max} - T_{min})$$

where L is a constant that depends on the values of $A, B, C, ω, φ$, and $β$.

The expression for $t_c$ in the LPPLS model is derived based on the assumption that the critical time lies within the interval $[T_{min}, T_{max}]$, and that the log-periodic power law pattern is symmetric around the critical time. The expression determines the location of the critical time based on the minimum and maximum values of time in the training data set and the value of the exponent m, which determines the rate of increase in the frequency of oscillations as the critical time is approached.

````python
from btc_analysis import *
from btc_data import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
import math
from scipy.optimize import minimize, Bounds, LinearConstraint
import lmfit

def load_data():
    # Load bitcoin prices data
    data = clean_and_transform_data(read_data("../datasets/public/btc.csv"), read_data("../datasets/public/btc_google_trend.csv"))
    prices = data['Price'].values

    # Rename the time column to date
    data = data.rename(columns={'time': 'date'})
    
    # Create new time column
    start_date = pd.to_datetime(data['date'].iloc[0]).date()
    data['time'] = (pd.to_datetime(data['date']).dt.date - start_date).dt.days + 1
    
    # Add ln_P column to dataframe
    data['ln_P'] = np.log(prices)
    data = data.dropna()
    
    data = data[['date', 'time', 'Price', 'ln_P']]
    return data

def split_data(data):
    # Split data into training and testing sets
    # Split data into training and testing sets
    split_idx = int(len(data) * 0.8)
    train_data = data[:split_idx]
    test_data = data[split_idx:]

    X_train = np.array(train_data['time'])
    X_test = np.array(test_data['time'])
    y_train = np.array(train_data['ln_P'])
    y_test = np.array(test_data['ln_P'])

    return X_train, X_test, y_train, y_test

# Print estimated parameters
def print_params(optimal_param):
    print('Estimated A: {:.4f}'.format(optimal_params[0]))
    print('Estimated B: {:.4f}'.format(optimal_params[1]))
    print('Estimated C: {:.4f}'.format(optimal_params[2]))
    print('Estimated w: {:.4f}'.format(optimal_params[3]))
    print('Estimated phi: {:.4f}'.format(optimal_params[4]))
    print('Estimated D: {:.4f}'.format(optimal_params[5]))
    print('Estimated beta: {:.4f}'.format(optimal_params[6]))
    print('Estimated m: {:.4f}'.format(optimal_params[7]))
    

def evaluate_model(X_test, y_test, P, A, B, C, w, phi, D, beta):
    # Evaluate model on testing data
    y_pred = compute_LPPLS(X_test, P, A, B, C, w, phi, D, beta)
    mse = np.mean((y_test - y_pred)**2)
    return mse, y_pred

def fill_LPPLS(data, A, B, C, w, phi, D, beta, m, t_c):
    """
    Given the LPPLS parameters and training data, predict the LPPLS values for the test data.
    """
    LPPLS = compute_LPPLS(data, A, B, C, w, phi, D, beta, m, t_c)
    return LPPLS


def fill_LPPLS_to_price(LPPLS):
    """
    Given the LPPLS parameters and training LPPLS values, predict the asset prices for the test data.
    """
    P = np.exp(LPPLS)
    return P

def compute_LPPLS(X, A, B, C, w, phi, D, beta, m, t_c):
    t = X
    LPPLS = A + B*(t_c - t)**m + C*(t_c - t)**m*np.cos(w*np.log(t_c - t) + phi) + D*np.sum(np.exp(-beta*(t_c - t)))
    return LPPLS

def lppls_model(params, X, t_c):
    A, B, C, w, phi, D, beta, m = params
    return compute_LPPLS(X, A, B, C, w, phi, D, beta, m, t_c)

def compute_t_c(X_train, m):
    t_c = X_train[-1] - ((m-1)/m)*(X_train[-1] - X_train[0])
    return t_c


def loss(params, X_train, y_train, lambda_reg):
    # Define loss function with L2 regularization
    A, B, C, w, phi, D, beta, m = params
    t_c = X_train[-1] - ((m-1)/m)*(X_train[-1] - X_train[0])
    y_pred = lppls_model(params, X_train, t_c)
    mse_loss = np.mean((y_train - y_pred)**2)
    l2_reg = lambda_reg * np.sum(params**2)
    return mse_loss + l2_reg


def estimate_optimal_params(X_train, y_train):
    # Estimate optimal parameters using training data with regularization
    
    lambda_reg = 0.1 # regularization strength
    
    # Minimize loss function with L2 regularization
    # Initial parameter values
    init_params = np.array([10.0, -10.0, 10.0, 10.0, 1.0, 0.01, 0.1, 1.5])

    bounds = Bounds([0, -np.inf, 0, 0, -np.pi/2, 0, 0, 1], np.inf)
    cons = LinearConstraint(np.array([[-1, 1, 1, 0, 0, 0, 0, -1/m], [0, 0, 0, -1, -1, 0, 0, 0]]), np.array([-np.inf, 0]), np.array([0, 0]))

    # Minimize loss function with L2 regularization
    res = minimize(loss, init_params, args=(np.array(X_train), np.array(y_train), lambda_reg), method='L-BFGS-B', bounds=bounds, constraints=cons)

    # Extract optimal parameters
    A_opt, B_opt, C_opt, w_opt, phi_opt, D_opt, beta_opt, m_opt = res.x

    return A_opt, B_opt, C_opt, w_opt, phi_opt, D_opt, beta_opt, m_opt
    
# Load data
data = load_data()
# data.columns
# Index(['date', 'time', 'Price', 'ln_P', 'LPPLS', 'LPPLS_Price'], dtype='object')

# Split data into training and testing sets
X_train, X_test, y_train, y_test = split_data(data)

# Estimate optimal parameters using training data with regularization
A, B, C, w, phi, D, beta, m = optimal_params = estimate_optimal_params(X_train, y_train)

# Fill in missing values in the "LPPLS" column of the dataframe using the LPPLS model
data['LPPLS'] = fill_LPPLS(data['time'], A, B, C, w, phi, D, beta, m, compute_t_c(np.array(X_train), m))

# Fill in missing values in the "LPPLS_Price" column of the dataframe using the LPPLS columnt
data['LPPLS_Price'] = fill_LPPLS_to_price(data['LPPLS'])

# Params
print_params(optimal_params)
````

````python
import numpy as np
import lmfit

# Define the LPPLS model function
def lppls_model(params_, t):
    A = params_['A'].value
    B = params_['B'].value
    C = params_['C'].value
    omega = params_['omega'].value
    phi = params_['phi'].value
    D = params_['D'].value
    beta = params_['beta'].value
    m = params_['m'].value
    tc = params_['tc'].value
    return A + B*(tc-t)**m + C*(tc-t)**m*np.cos(omega*np.log(tc-t)+phi) + D*np.sum(np.exp(-beta*(t-tc)))

# Define the function for calculating the residual sum of squares
def residual(params_, t, data):
    return lppls_model(params_, t) - data

# Define the function for calculating the critical time
def critical_time(params_, tmin, tmax):
    m = params_['m'].value
    beta = params_['beta'].value
    D = params_['D'].value
    tc = tmax - (tmax-tmin)*np.exp((-np.log(beta*D))/(m-1))
    return tc

# Generate some example data
t = np.arange(0, 20, 0.1)
P = np.exp(np.sin(t))

params_ = lmfit.Parameters()
params_.add('A', value=1.0, min=-np.inf, max=np.inf)
params_.add('B', value=1.0, min=-np.inf, max=np.inf)
params_.add('C', value=1.0, min=-np.inf, max=np.inf)
params_.add('omega', value=5.0, min=0.0, max=np.inf)
params_.add('phi', value=0.0, min=-np.pi, max=np.pi)
params_.add('D', value=0.1, min=0.0, max=np.inf)
params_.add('beta', value=0.1, min=0.0, max=np.inf)
params_.add('m', value=0.5, min=0.0, max=2.0)
params_.add('tc', value=10.0, min=0.0, max=np.max(t))

model = lmfit.Model(lppls_model)

# Perform the nonlinear regression analysis
result = model.fit(P, params_, t=t, method='leastsq')

# Estimate the critical time
tc = critical_time(result.params_, np.min(t), np.max(t))
print('Estimated critical time: %.4f' % tc)

# Generate a plot of the LPPLS model curve and visualize the fit of the model to the input data
import matplotlib.pyplot as plt
plt.plot(t, P, 'bo', label='data')
plt.plot(t, result.best_fit, 'r-', label='fit')
plt.legend()
plt.show()
````

````python
from scipy.signal import find_peaks
import matplotlib.pyplot as plt

def plot_all_bubbles(data=data, distance=100, prominence=1):
    peaks, _ = find_peaks(data['Price'], distance=distance, prominence=prominence)

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

    # Plot all bubbles in the first subplot
    ax1 = axes[0]
    ax1.plot(data['date'], data['Price'])
    ax1.scatter(data['date'][peaks], data['Price'][peaks], color='red')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Price')
    ax1.set_title('Price Bubbles with Peaks Highlighted')

    # Plot all ln bubbles in the second subplot
    ax2 = axes[1]
    ax2.plot(data['date'], data['ln_P'], color="green")
    ax2.scatter(data['date'][peaks], data['ln_P'][peaks], color='red')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('ln_P')
    ax2.set_title('ln_P Bubbles with Peaks Highlighted')

    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
    
    plt.tight_layout()
    plt.show()

    return peaks    
    
def plot_bubbles(data=data, 
                distance=100,
                prominence=1,
                subplot_width=4,
                subplot_height=3, 
                subplot_cols=3):

    # Find peak prices with a distance of 200 and a prominence of 100
    peaks, _ = find_peaks(data['Price'], distance=distance, prominence=prominence)

    # Create subplots for each bubble
    num_bubbles = len(peaks)
    subplot_rows = (num_bubbles - 1) // subplot_cols + 1  # Calculate number of subplot rows based on number of bubbles and subplot columns
    fig, axes = plt.subplots(nrows=subplot_rows, ncols=subplot_cols, figsize=(subplot_width*subplot_cols, subplot_height*subplot_rows))

    for i, peak_index in enumerate(peaks):
        # Define the date range for the bubble
        start_index = max(0, peak_index-distance)
        end_index = min(len(data)-1, peak_index+distance)

        # Subset the data to the bubble range
        bubble_data = data.iloc[start_index:end_index+1].reset_index(drop=True)

        # Find the highest peak in the bubble
        max_peak_index = bubble_data['Price'].idxmax()

        # Calculate subplot coordinates
        subplot_row = i // subplot_cols
        subplot_col = i % subplot_cols

        # Plot the data with the highest peak highlighted in red
        ax = axes[subplot_row, subplot_col]
        ax.plot(bubble_data['date'], bubble_data['Price'])
        ax.scatter(bubble_data['date'][max_peak_index], bubble_data['Price'][max_peak_index], color='red')
        ax.set_xlabel('Date')
        ax.set_ylabel('Price')
        ax.set_title('Bubble {} - Peak Price: {}'.format(i+1, bubble_data['Price'][max_peak_index]))

        plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
    # Remove unused subplots
    for i in range(num_bubbles, subplot_rows*subplot_cols):
        fig.delaxes(axes[i // subplot_cols, i % subplot_cols])

    plt.tight_layout()
    plt.show()
    
    return peaks

def plot_ln_bubbles(peaks,
                    color="green",
                    data=data, 
                    distance=100,
                    prominence=1,
                    subplot_width=4,
                    subplot_height=3, 
                    subplot_cols=3):

    # Create subplots for each bubble
    num_bubbles = len(peaks)
    subplot_rows = (num_bubbles - 1) // subplot_cols + 1  # Calculate number of subplot rows based on number of bubbles and subplot columns
    fig, axes = plt.subplots(nrows=subplot_rows, ncols=subplot_cols, figsize=(subplot_width*subplot_cols, subplot_height*subplot_rows))
    sub_bubbles_slices = []
    for i, peak_index in enumerate(peaks):
        # Define the date range for the bubble
        start_index = max(0, peak_index-distance)
        end_index = min(len(data)-1, peak_index+distance)

        # Subset the data to the bubble range
        bubble_data = data.iloc[start_index:end_index+1].reset_index(drop=True)
        
        sub_bubbles_slices.append(data.iloc[start_index:end_index+1].reset_index(drop=True))
        # Find the highest peak in the bubble
        max_peak_index = bubble_data['ln_P'].idxmax()

        # Calculate subplot coordinates
        subplot_row = i // subplot_cols
        subplot_col = i % subplot_cols

        # Plot the data with the highest peak highlighted in red
        ax = axes[subplot_row, subplot_col]
        ax.plot(bubble_data['date'], bubble_data['ln_P'], color=color)
        ax.scatter(bubble_data['date'][max_peak_index], bubble_data['ln_P'][max_peak_index], color='red')
        ax.set_xlabel('Date')
        ax.set_ylabel('ln_P')
        ax.set_title('Bubble {} - Peak Price: {}'.format(i+1, bubble_data['ln_P'][max_peak_index]))

        plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
    # Remove unused subplots
    for i in range(num_bubbles, subplot_rows*subplot_cols):
        fig.delaxes(axes[i // subplot_cols, i % subplot_cols])

    plt.tight_layout()
    plt.show()
    
    return peaks, sub_bubbles_slices
````
