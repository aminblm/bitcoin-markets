# Combining Models for Accurate Forecasting of Bitcoin Market Prices: A Novel Approach

Boulouma A., 2023

## Abstract

The rapid growth of cryptocurrency markets has created a need for accurate forecasting of Bitcoin prices, which can have significant impacts on global financial systems. In this study, we aim to determine the combination of models that yields the highest predictive power for forecasting Bitcoin market prices. Our research question is whether combining multiple models can lead to more accurate predictions than individual models.

Our proposed research is of significant importance as it provides a novel approach to forecasting Bitcoin market prices, which can inform investment decisions and help mitigate financial risks. We aim to utilize several well-known models, including the Jump-diffusion models, Dynamic Stochastic General Equilibrium (DSGE) Models, Bayesian Time Series Analysis, Empirical Asset Pricing, as well as Robert Shiller's theories of Speculative Asset Prices and Narrative Economics.

Our research methodology involves the use of historical data on Bitcoin prices, as well as statistical analysis and machine learning techniques to determine the most effective combination of models. We will compare the predictive power of combined models with that of individual models using metrics such as Mean Absolute Error (MAE) and Root Mean Square Error (RMSE).

The expected outcomes of our research are the identification of the optimal combination of models for accurate forecasting of Bitcoin market prices, which can inform financial decision-making and contribute to the advancement of cryptocurrency research. Our study will also provide insights into the effectiveness of combining different models for forecasting other financial assets.

## Introduction

Over the past decade, the world has witnessed a proliferation of cryptocurrencies, and Bitcoin has emerged as the undisputed leader in this rapidly evolving market. The growth of Bitcoin has attracted the attention of investors, economists, and financial analysts worldwide, as its volatile price fluctuations have sparked debates about the potential of this asset class. The complex nature of Bitcoin, coupled with its decentralized and unregulated framework, makes forecasting the future of Bitcoin prices an arduous task. The need for accurate forecasting has become increasingly essential as the financial industry and policymakers grapple with the challenges posed by cryptocurrencies.

Accurate forecasting of Bitcoin prices is vital because it provides insights into market trends and enables traders to make informed decisions. Forecasting models can help predict future prices by analyzing past prices and identifying patterns and trends in the market. These models have become increasingly popular in the financial industry, and several forecasting models have been developed to predict Bitcoin prices. However, no model has been able to achieve high predictive accuracy consistently, and the search for a reliable model remains a challenging and pressing research question.

In this context, the importance of identifying the best combination of models that yields the highest predictive power for forecasting Bitcoin market prices cannot be overstated. The Nobel Prize-winning research works in finance, such as Jump-diffusion models, Dynamic Stochastic General Equilibrium (DSGE) Models, Bayesian Time Series Analysis, and Empirical Asset Pricing, provide a theoretical foundation for modeling financial markets. These models offer insights into the behavior of financial markets and can be used to forecast future prices. The recent works of Robert Shiller, such as "Speculative Asset Prices" and "Narrative Economics," have provided new perspectives on how human behavior influences financial markets and have opened new avenues for research.

Therefore, this proposal aims to investigate the effectiveness of different combinations of models in predicting Bitcoin market prices accurately. The research will draw on the theoretical foundations provided by the Nobel Prize-winning research works and other relevant academic literature to construct various models. The methodology will involve testing different combinations of these models to determine which combination yields the highest predictive accuracy. The expected outcomes will provide insights into the best combination of models that can be used to forecast Bitcoin market prices accurately. This research will provide valuable inputs for traders, policymakers, and investors who seek to make informed decisions based on accurate price predictions.
