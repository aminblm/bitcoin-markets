## Abstract

The rapid growth of cryptocurrency markets has created a need for accurate forecasting of Bitcoin prices, which can have significant impacts on global financial systems. In this study, we aim to determine the combination of models that yields the highest predictive power for forecasting Bitcoin market prices. Our research question is whether combining multiple models can lead to more accurate predictions than individual models.

Our proposed research is of significant importance as it provides a novel approach to forecasting Bitcoin market prices, which can inform investment decisions and help mitigate financial risks. We aim to utilize several well-known models, including the Jump-diffusion models, Dynamic Stochastic General Equilibrium (DSGE) Models, Bayesian Time Series Analysis, Empirical Asset Pricing, as well as Robert Shiller's theories of Speculative Asset Prices and Narrative Economics.

Our research methodology involves the use of historical data on Bitcoin prices, as well as statistical analysis and machine learning techniques to determine the most effective combination of models. We will compare the predictive power of combined models with that of individual models using metrics such as Mean Absolute Error (MAE) and Root Mean Square Error (RMSE).

The expected outcomes of our research are the identification of the optimal combination of models for accurate forecasting of Bitcoin market prices, which can inform financial decision-making and contribute to the advancement of cryptocurrency research. Our study will also provide insights into the effectiveness of combining different models for forecasting other financial assets.

### Overview of the models used in the study

In this study, we employ various models to analyze and forecast Bitcoin prices. These models include the Jump-diffusion model, the Dynamic Stochastic General Equilibrium (DSGE) model, Bayesian Time Series Analysis, Empirical Asset Pricing models, Narrative Economics, and Speculative Asset Prices.

The DSGE model, developed by Finn Kydland and Edward Prescott in 1997, is a macroeconomic model that describes the behavior of the economy as a whole. It assumes that economic agents are rational and forward-looking and that economic fluctuations are caused by various shocks. 

Bayesian Time Series Analysis, introduced by Thomas Sargent in 2008, is a statistical approach that uses Bayesian inference to estimate the parameters of a time series model. This method allows us to incorporate prior knowledge into the model and update the parameters as new data becomes available. Bayesian Time Series Analysis has gained popularity in forecasting Bitcoin prices due to its ability to handle complex time series data and provide reliable estimates of the model parameters.

Empirical Asset Pricing models, developed by Eugene Fama and Lars Peter Hansen in 2011, analyze the relationship between risk and return in financial markets. These models use various factors, such as market capitalization, book-to-market ratio, and momentum, to explain the variation in asset returns. Empirical Asset Pricing models have been applied to analyze the risk and return of Bitcoin and have been found to explain a significant portion of the variation in Bitcoin returns.

Finally, Narrative Economics and Speculative Asset Prices, introduced by Robert Shiller in 2017 and 2005, respectively, are approaches that focus on the role of narratives and psychology in financial markets. Narrative Economics uses data from sources such as social media and news articles to analyze the stories and narratives that influence market behavior. Speculative Asset Prices, on the other hand, focus on the psychological factors that drive market bubbles and crashes. These models have been applied to analyze the behavior of Bitcoin prices and have provided insights into the role of narratives and psychology in Bitcoin market dynamics.

In summary, this study employs a range of models, from traditional finance models such as the Jump-diffusion model and Empirical Asset Pricing models to more recent approaches such as Bayesian Time Series Analysis, Narrative Economics, and Speculative Asset Prices. These models provide a comprehensive understanding of the dynamics of Bitcoin prices and enable us to forecast future market trends with greater accuracy.

### 1976, Jump-diffusion models, Robert Merton (Nobel Prize 1997) 

Jump diffusion models for Bitcoin prices forecasting can be formulated as follows:

Let $S_t$ be the Bitcoin price at time $t$, and let $\mu$ and $\sigma$ be the constant drift and volatility of the price process, respectively. Additionally, let $\lambda$ and $\delta$ be the intensity and magnitude of the jump process, respectively.

Then, the stochastic process $S_t$ can be modeled as a jump diffusion process given by the following stochastic differential equation:

\begin{equation}
    dS_t = \mu S_t dt + \sigma S_t dW_t + S_t dJ_t
\end{equation}

where $W_t$ is a standard Wiener process and $J_t$ is a Poisson process with jump sizes distributed according to a stable distribution with location parameter $\delta$ and scale parameter $\lambda$.

The parameters of the model, $\mu$, $\sigma$, $\lambda$, and $\delta$, can be estimated using maximum likelihood estimation or other statistical methods. The resulting model can then be used to forecast future Bitcoin prices and estimate the associated uncertainty.

In this formulation, the variables are defined as follows:

- $S_t$: the Bitcoin price at time $t$
- $\mu$: the constant drift of the price process
- $\sigma$: the constant volatility of the price process
- $\lambda$: the intensity of the jump process
- $\delta$: the magnitude of the jump process
- $W_t$: a standard Wiener process

The Bayesian Time Series Analysis model is defined as follows:

$$y_t = \beta_0 + \beta_1 x_t + \epsilon_t$$
$$\epsilon_t \sim \mathcal{N}(0, \sigma^2)$$

where $y_t$ represents the Bitcoin price at time $t$, $x_t$ represents the relevant economic indicator at time $t$, $\beta_0$ and $\beta_1$ are the intercept and slope coefficients, and $\epsilon_t$ is the error term. The error term is assumed to follow a normal distribution with mean 0 and variance $\sigma^2$.

Bayesian Time Series Analysis requires that prior distributions be specified for the model parameters. In this case, we will use normal distributions for the intercept and slope coefficients:

$$\beta_0 \sim \mathcal{N}(0, 10^2)$$
$$\beta_1 \sim \mathcal{N}(0, 10^2)$$

The posterior distributions for the model parameters can be estimated using Bayesian inference. This involves updating the prior distributions based on the observed data to obtain the posterior distributions:

$$p(\beta_0, \beta_1|\textbf{y}, \textbf{x}) \propto p(\textbf{y}|\beta_0, \beta_1, \textbf{x})p(\beta_0)p(\beta_1)$$

where $\textbf{y}$ represents the vector of Bitcoin prices, $\textbf{x}$ represents the matrix of economic indicators, and $p(\textbf{y}|\beta_0, \beta_1, \textbf{x})$ is the likelihood function.

Once the posterior distributions for the model parameters have been estimated, they can be used to make predictions for future Bitcoin prices. This involves simulating from the posterior distributions to obtain a distribution of future Bitcoin prices:

$$y_{t+1} \sim \mathcal{N}(\beta_0 + \beta_1 x_{t+1}, \sigma^2)$$

where $y_{t+1}$ represents the predicted Bitcoin price at time $t+1$ and $x_{t+1}$ represents the relevant economic indicator at time $t+1$.
